```python
# 📊 Summary and Visualization Notebook

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from GitHub
url = 'https://raw.githubusercontent.com/estherpraveena/data-analyst-esther/main/data/rental-standards-current-issues.csv'
df = pd.read_csv(url, delimiter=';')

# Strip spaces from column names
df.columns = df.columns.str.strip()

# 👑 Top 10 Business Operators by Number of Violations
top_operators = df['BUSINESSOPERATOR'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_operators.values, y=top_operators.index, palette='magma')
plt.title('Top 10 Business Operators by Violations')
plt.xlabel('Number of Violations')
plt.ylabel('Business Operator')
plt.tight_layout()
plt.show()

# 🌎 Top 10 Areas by Number of Violations
top_areas = df['Geo Local Area'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_areas.values, y=top_areas.index, palette='coolwarm')
plt.title('Top 10 Areas by Violations')
plt.xlabel('Number of Violations')
plt.ylabel('Geo Local Area')
plt.tight_layout()
plt.show()

# 📋 Summary Table: Average Outstanding by Area
summary = df.groupby('Geo Local Area')['TOTALOUTSTANDING'].mean().sort_values(ascending=False)
summary.head(10)
```
